```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Load Iris dataset from sklearn
try:
    iris = load_iris()
    df = pd.DataFrame(iris.data, columns=iris.feature_names)
    df['species'] = iris.target
    df['species'] = df['species'].map(dict(zip(range(3), iris.target_names)))

    print("Dataset loaded successfully.")
except Exception as e:
    print(f"Error loading dataset: {e}")
```

```python
# Display the first few rows
print("\n First 5 rows of the dataset:")
print(df.head())

# Explore the structure of the dataset
print("\n Dataset Info:")
print(df.info())

# Check for missing values
print("\n Missing values per column:")
print(df.isnull().sum())

df.dropna(inplace=True)
```

```python
# Basic statistics
print("\n Basic statistics:")
print(df.describe())

# Group by species and compute mean
grouped_mean = df.groupby('species').mean()
print("\n Mean of each numerical column grouped by species:")
print(grouped_mean)

# Observation
print("\n Observations:")
print("- Iris-virginica generally has the highest petal length and width.")
print("- Iris-setosa has the lowest petal measurements.")
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 5))
plt.plot(df.index, df['sepal length (cm)'], label='Sepal Length', color='green')
plt.title('Line Chart of Sepal Length over Entries')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.legend()
plt.show()

```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='species', y='petal length (cm)', data=df, hue='species', palette='Set2', legend=False)
plt.title('Average Petal Length per Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
# Histogram: Distribution of Sepal Width
plt.figure(figsize=(8, 5))
plt.hist(df['sepal width (cm)'], bins=15, color='skyblue', edgecolor='black')
plt.title('Histogram of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter Plot: Sepal Length vs Petal Length
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal length (cm)', y='petal length (cm)', hue='species', palette='Set1')
plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.show()
```
